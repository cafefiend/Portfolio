```python
import pandas as pd
amk_hdb = pd.read_csv ("MedianResalePricesforRegisteredApplicationsbyTownandFlatType.csv")
amk_hdb.head (10)

```

```python
amk = amk_hdb[amk_hdb['town'].isin(['Ang Mo Kio', 'ANG MO KIO'])]
amk.head(10)
```

```python
amk['town'] = amk['town'].replace('ANG MO KIO', 'Ang Mo Kio')
amk.tail (50)
```

```python
amk.head (100)

```

```python
print (amk['price'].dtype)
print (amk ['quarter'].dtype)
```

```python
amk['quarter'] = pd.PeriodIndex(amk['quarter'], freq='Q').to_timestamp()
print(amk['quarter'].dtype)  # Should show datetime64[ns]
print(amk.head(50))  # Check the first few rows
```

```python
amk_2024 = amk[amk['quarter'].dt.year == 2024]
print(amk_2024)
```

```python
print(amk['quarter'].dtype)  # Should show datetime64[ns]
print (amk['price'].dtype)
print (amk.tail (50))
```

```python
amk_hdb_type = amk[~amk['flat_type'].isin(['1-room', '2-room', 'executive', 'Executive', 'EXECUTIVE'])]
amk_hdb_type.head (50)
```

```python
amk_hdb_type.tail (50)
```

```python
print (amk_hdb_type['price'].dtype)
```

```python
amk_hdb_type['price'] = pd.to_numeric(amk_hdb_type['price'], errors='coerce')
print(amk_hdb_type.dtypes) 
print (amk_hdb_type.head (50))
```

```python
print (amk_hdb_type.tail (50))
```

```python
aggregated_median_with_nan = amk_hdb_type.groupby(['quarter', 'town', 'flat_type']).agg(median_price=('price', lambda x: x.dropna().median())  # Drops NaNs only for median calculation
).reset_index()
print(aggregated_median_with_nan.head())
```

```python
print(aggregated_median_with_nan.tail(50))
```

```python
aggregated_median_with_nan.to_csv ("AMK_aggregated_median.csv", index=False)
```

```python
import pandas as pd
aggregated_median_with_nan = pd.read_csv ("AMK_aggregated_median.csv")
aggregated_median_with_nan.head (30)
```

```python
aggregated_median_with_nan = aggregated_median_with_nan.dropna(subset=['median_price'])
aggregated_median_with_nan['quarter'] = pd.to_datetime(aggregated_median_with_nan['quarter'])
yearly_price_by_flat_type = aggregated_median_with_nan.groupby(['year', 'flat_type'], as_index=False)['median_price'].median()
yearly_price_by_flat_type.rename(columns={'median_price': 'agg_median'}, inplace=True)
print(yearly_price_by_flat_type)

```
